# G&L Gyro - Summer 2019 Data Analysis

## 1. Introduction

"G&L Gyro" is a family-run food business specializing in serving fresh Greek food and other barbequed foods at fairs and other events across Long Island. Below we will analyze data from one employee of the stand from the Summer 2019 season. Let's load the .csv file and see what we are working with.

```python
#Import module
import pandas as pd

#Read in .csv file of data
data = pd.read_csv('Downloads/2019 GL Gyro Money Made - Sheet1 (2).csv')

#Print out all data
pd.set_option('display.max_rows', data.shape[0]+1)
print(data)
```

```python
#Print out the head (first five values)
data.head()
```

## 2. Analysis of Total Amount Earned throughout the entire season

Now with the data in our hands, let's see how much money was earned as the season went on. With this job, most of the "better" jobs (in terms of profit) are predicted to be in the deep summer months (June, July, August), as this is the prime season for fairs and summer entertainment events. These months are in Job # ranges 19-50 (represented by the index in the csv file). We can use a simple line graph to try and confirm this. 

```python
#Import matplotlib library for the plotting
import matplotlib.pyplot as plt

#Set parameters for graph and display
data.plot(y='Total($)', use_index = True)
plt.xlabel('Job #')
plt.ylabel('Amount Earned($)')
plt.rcParams["figure.figsize"] = (20, 6)
plt.xticks(range(0, 57))
plt.show()
```

So we can see by this graph that, with the exception of Job #2 and #16, the best earnings were indeed in the deeper summer months. We have high spikes in the graph from Job #28 (July 4th) and on. Since the employees got paid a flat rate of $100 a job + any additional tips from customers, the high values really function to show us how busy a certain job was. More people at an event = more tips. Of course, this isn't always true. There could be jobs with not too many people but one customer gives a very high tip (that would equate to what 5 or 6 customers combined would usually give). But the more rewarding jobs do seem to be located in the heart of the season.

Below are some straightforward statistical measures for the Total column.

```python
print('Total Money Earned:$ 7739')
print('Mean Amount (per job): ', data['Total($)'].mean())
print('Median: ', data['Total($)'].median())
print('Standard Deviation: ', data['Total($)'].std())
```

## 3. Tips by Location (who's the most generous?)

Some locations on Long Island are notorious for not being very tip-friendly and on the other hand, employees end up making more than the flat rate of pay at other locations. Seasoned G&L employees know what jobs pay the best with tips and which one's don't. Again, as stated before, the job outlook sometimes varies from year to year--usually dependent on how enjoyable the events are to people as the years go by. Let's take a look at who was the most generous for the 2019 season.

```python
#Imports
import numpy as np

#Set bar graph parameters
plt.bar(x = np.arange(0, 58), height=data['Tips($)'], color='green')
plt.title("Tips and Locations")
plt.xticks(np.arange(0,58), data['Place'], rotation=90)
plt.xlabel("Location")
plt.ylabel("Tips Earned per Job ($)")
plt.rcParams["figure.figsize"] = (20, 20)
plt.show()
```

GSBMF(The Great South Bay Music Festival), located in Patchogue, ended up being the most tip-friendly with over 70 dollars in tips. Keep in mind, this is 70 dollars for each person working that job. Using the .csv file, we can see that there was a total of 3 employees getting the tips that night, totaling to over 200 dollars in tips.

Other locations here had a solid cashout on tips each day or night of the job, like the Hamptons for example, which banked about 20 dollars each job. 

Some low-tipping jobs include Broadway Mall, Ampitheater (weekend 1), Astoria, and Bayville.

Below are some simple statistical measures for the Tips column.

```python
print('Total Tips Earned:$', data['Tips($)'].sum())
print('Mean Amount (per job): ', data['Tips($)'].mean())
print('Median: ', data['Tips($)'].median())
print('Standard Deviation: ', data['Tips($)'].std())
```

## 4. Employee Counter

Now it's time to hit the last column of this dataset--the people. We will get a count on how many times this particular employee (the creator of the dataset) has worked with every other employee.

```python
peopleArray = data['Worked With']
print(*peopleArray, sep = ', ')
```

Above here is a list containing a tally on each time someone worked a job with our employee. Let's get a count on each name.

```python
#Create one big string from the list to perform operations on
bigString = ' '.join(peopleArray)

#Simple method to count the names
def countNames (inputName):
    toReturn = bigString.count(inputName)
    print(inputName, toReturn)

#Call the method on all the names
countNames('Andrew')
countNames('George')
countNames('Lea')
countNames('Danny')
countNames('Squid')
countNames('Ando')
countNames('Paul')
countNames('Grace')
countNames('Jay')
countNames('Tracy')
countNames('Mikayla')
countNames('Kelly')
countNames('Dean')
countNames('Marisa')
countNames('Adna')
countNames('Brooke')
countNames('Kerri')
countNames('Casey')
countNames('Christian')
countNames('Chichito')
countNames('Frucci')
countNames('Amazon')
countNames('Carrie')
countNames('Shannon')
countNames('Juan')
countNames('Adam')
countNames('Kim')
countNames('Sara')
countNames('Vic')
countNames('Lori')
countNames('Kiki')
countNames('Alexis')
countNames('Anthony')
countNames('Matt')
countNames('Phil')
```
